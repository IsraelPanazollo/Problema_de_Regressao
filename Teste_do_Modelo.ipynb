# Teste do modelo de regressão desenvolvido

```python
import pandas as pd
import numpy as np
```

## Instanciação do modelo

```python
import pickle

modelo = open('modelo','rb')
RFR = pickle.load(modelo)
modelo.close()
```

## Informações do cliente

```python
cliente_fulano = pd.DataFrame({'tax_iptu':[770],
                              'tax_condo':[740],
                              'quartos':[0],
                              'banheiros':[1],
                              'suites':[0],
                              'garagens':[1],
                              'area_usavel':[60],
                              'area_total':[60],
                              'pisos':[4],
                              'unidades_sobre_piso':[25],
                              'unidades_piso':[2],
                              'tipo_Casa':[0],
                              'tipo_Outros':[1],
                              'zona_Zona Central':[0],
                              'zona_Zona Norte':[0],
                              'zona_Zona Oeste':[1],
                              'zona_Zona Sul':[0]})
cliente_fulano
```

```python
dados_cliente = np.log1p(cliente_fulano)
dados_cliente
```

## Aplicação do modelo e previsão do preço do imóvel

```python
predicao_cliente_fulano = RFR.predict(dados_cliente)
print(f" O imóvel do Fulano com essas características custaria: {np.expm1(predicao_cliente_fulano).round(2)}")
```
